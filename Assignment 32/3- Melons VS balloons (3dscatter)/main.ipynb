```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 100
melons = pd.DataFrame({'length' : np.random.normal(18.5, 3, N),
                      'width' : np.random.normal(18.5, 3, N),
                      'weight' : np.random.normal(2700, 100, N)})

balloons = pd.DataFrame({'length' : np.random.normal(27, 2, N),
                         'width' : np.random.normal(17.5, 2.5, N),
                         'weight' : np.random.normal(35, 5, N)})
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(projection='3d')
ax.scatter(melons['width'], melons['length'], melons['weight'], c='blue', marker='o', label='Melons')
ax.scatter(balloons['width'], balloons['length'], balloons['weight'], c='red', marker='^', label='Balloons')

ax.set_title('Melons VS Balloons')
ax.set_xlabel('Width')
ax.set_ylabel('Length')
ax.set_zlabel('Weight')
ax.legend()

plt.show()
```
