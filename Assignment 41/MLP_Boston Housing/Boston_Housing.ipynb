```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
import numpy as np
import wandb
from wandb.keras import WandbCallback
```

```python
wandb.init(project="Boston Housing")
```

```python
config = wandb.config
config.learning_rate = 0.001
```

### Data

```python
dataset = tf.keras.datasets.boston_housing
(X_train, Y_train), (X_test, Y_test) = dataset.load_data()
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
print(X_test.shape)
print(Y_test.shape)
```

### Model

```python
model = tf.keras.models.Sequential([
    Dense(512, input_dim=13, activation="relu"),
    Dense(256, activation="relu"),
    # Dense(64, activation="relu"),
    Dense(1, activation="linear")
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=config.learning_rate),
              loss=tf.keras.losses.MeanSquaredError())
```

```python
model.summary()
```

### Train

```python
model.fit(X_train, Y_train, epochs=350, callbacks=[WandbCallback()])
```

```python
model.evaluate(X_test, Y_test)
```

### Inference

```python
my_house = np.array([[0.00632, 18.0, 2.31, 0.0, 0.532, 6.575, 65.2, 4.09, 1, 296.0, 15.3, 396.9, 4.98]])

model.predict(my_house)
```
